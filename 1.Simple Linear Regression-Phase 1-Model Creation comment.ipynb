```python
import pandas as pd
#import pandas python datastructures and operations and assign as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#call the read_csv function to read the Salary_Data.csv and assign in dataset variable
```

```python
dataset
#Run the dataset variable and fetch data's present in the table
```

```python
independent=dataset[["YearsExperience"]] #Assign the yearsexperience as input on independent variable
independent
```

```python
independent=dataset[["YearsExperience"]]
independent
```

```python
dependent=dataset[["Salary"]] #Assign the salary as output on dependent variable
dependent
```

```python
from sklearn.model_selection import train_test_split #import train_test_split function in model function class from sklearn library
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) 
# split input and output as training set, test set randomly and assign in the variable
```

```python
from sklearn.linear_model import LinearRegression #import LinearRegression function in linear_model class from sklearn library 
regressor=LinearRegression() # Assign the LinearRegression function in regressor variable
regressor.fit(X_train,Y_train) # run the regressor fn to execute the linearreegression using regressor.fit
```

```python
weight=regressor.coef_ #call the regressor.coef_ function to find the slope and assign in the weight variable
weight #run the weight variable to find the slope value
```

```python
bais=regressor.intercept_ #call the regressor.intercept_ function to find the intercept and assign in the bais variable
bais #run the bais variable to find the bais value
```

```python
Y_pred=regressor.predict(X_test) 
#to execute the test set call the regressor.predict assign the X_test parameter and assign in the Y_pred variable
```

```python
from sklearn.metrics import r2_score #import the r2_score function in metrics class from sklearn library
r_score=r2_score(Y_test,Y_pred) #Assign the parameter in r2_score function to find the r2 value and assign to r_score variable
```

```python
r_score #run the r_score variable
```

```python
import pickle # import the pickle module to convert and stored objects into bytes
filename="finalized_model_linear.sav" #create the filename with .sav extention and assign in filename varible
```

```python
pickle.dump(regressor,open(filename,'wb')) #open the regressor variable and write bytes in filename using dump class
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#load the model using pickle.load class and rb, saved as loaded_model variable
```

```python
result=loaded_model.predict([[13]])
#execute the model using loaded_model.predict and 13 years experience assaign in result variable
```
